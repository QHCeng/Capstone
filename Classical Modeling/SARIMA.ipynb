### Stationarity and Seasonality Check

```python
# --- SARIMAX EDA: Stationarity and Seasonality ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller, kpss

def stationarity_tests(series, cluster_id):
    s = series.dropna()
    adf_p = adfuller(s)[1]
    try:
        kpss_p = kpss(s, regression='c')[1]
    except:
        kpss_p = np.nan

    print(f"\nCluster {cluster_id}")
    print(f"ADF p-value:   {adf_p:.4f}  →  {'Stationary' if adf_p<0.05 else 'Non-stationary'}")
    print(f"KPSS p-value:  {kpss_p:.4f}  →  {'Trend-stationary' if kpss_p>0.05 else 'Has trend'}")

# Apply to each cluster
for cid, g in cluster_monthly.groupby("Cluster"):
    s = g.set_index("date").asfreq("M")["total_amount"]
    stationarity_tests(s, cid)
```

### ACF/PACF (original & differenced)

```python
# --- SARIMAX EDA: ACF/PACF per cluster (raw and differenced) ---


M = 12  # monthly seasonality

def infer_d_D(s):
    # simple heuristic based on your earlier tests
    from statsmodels.tsa.stattools import adfuller
    def adf_p(x):
        try: return adfuller(x.dropna(), autolag="AIC")[1]
        except: return np.nan
    p_raw, p_d1, p_seas = adf_p(s), adf_p(s.diff()), adf_p(s.diff(M))
    d = 0 if p_raw < 0.05 else (1 if p_d1 < 0.05 else 2)
    D = 1 if p_seas >= 0.05 else 0
    return d, D

for cid, g in cluster_monthly.groupby("Cluster"):
    s = g.set_index("date").asfreq("M")["total_amount"].astype(float)

    # (1) RAW SERIES ACF/PACF
    fig, axes = plt.subplots(1, 2, figsize=(10, 3))
    plot_acf(s.dropna(), lags=48, ax=axes[0])
    axes[0].set_title(f"Cluster {cid} — ACF (raw)")
    plot_pacf(s.dropna(), lags=48, ax=axes[1], method="ywm")
    axes[1].set_title(f"Cluster {cid} — PACF (raw)")
    plt.tight_layout(); plt.show()

    # (2) DIFFERENCED SERIES ACF/PACF (apply d and D you’ll use in SARIMA)
    d, D = infer_d_D(s)
    sd = s.copy()
    if d == 1:  sd = sd.diff()
    if d == 2:  sd = sd.diff().diff()
    if D == 1:  sd = sd.diff(M)
    sd = sd.dropna()

    fig, axes = plt.subplots(1, 2, figsize=(10, 3))
    plot_acf(sd, lags=48, ax=axes[0])
    axes[0].set_title(f"Cluster {cid} — ACF (after d={d}, D={D})")
    plot_pacf(sd, lags=48, ax=axes[1], method="ywm")
    axes[1].set_title(f"Cluster {cid} — PACF (after d={d}, D={D})")
    plt.tight_layout(); plt.show()
```

### STL decomposition 

```python

M = 12  # monthly seasonality

def strength_of(component, remainder):
    """Hyndman-style strength: 1 - Var(R) / Var(R + C).
       Higher = stronger contribution from the component."""
    r = remainder.dropna().values
    c = component.reindex(remainder.index).values
    num = np.var(r, ddof=1)
    den = np.var(r + c, ddof=1)
    return np.nan if den == 0 else max(0.0, 1.0 - num/den)

for cid, g in sorted(cluster_monthly.groupby("Cluster"), key=lambda x: x[0]):
    s = (g.set_index("date")
           .asfreq("M")["total_amount"]
           .astype(float))

    # STL with robust fitting to reduce outlier influence
    stl = STL(s, period=M, robust=True)
    res = stl.fit()

    # compute strengths
    seas_strength  = strength_of(res.seasonal,  res.resid)
    trend_strength = strength_of(res.trend - res.trend.rolling(M, min_periods=1).mean(), res.resid)

    print(f"\nCluster {cid} — STL strengths")
    print(f"  Seasonal strength: {seas_strength:.3f}  (closer to 1 = strong annual cycle)")
    print(f"  Trend strength:    {trend_strength:.3f}  (closer to 1 = strong trend)")

    # plot
    fig = res.plot()
    fig.set_size_inches(10, 8)
    plt.suptitle(f"Cluster {cid} — STL Decomposition (period={M})", y=0.98)
    plt.tight_layout()
    plt.show()
```

### SARIMA for SINGLE CLUSTER

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller

def adf_p(s):
    try: return adfuller(s.dropna(), autolag='AIC')[1]
    except: return np.nan

def metrics(y_true, y_pred):
    y_true, y_pred = np.asarray(y_true), np.asarray(y_pred)
    mae  = np.mean(np.abs(y_true - y_pred))
    rmse = np.sqrt(np.mean((y_true - y_pred)**2))
    mape = np.mean(np.abs((y_true - y_pred)/(y_true + 1e-12))) * 100
    smape= 100*np.mean(2*np.abs(y_pred - y_true)/(np.abs(y_true)+np.abs(y_pred)+1e-12))
    return {"MAE":mae, "RMSE":rmse, "MAPE%":mape, "sMAPE%":smape}

# pick a cluster id that exists in your data (e.g., 0)
cid = cluster_monthly['Cluster'].iloc[0]
sub = (cluster_monthly[cluster_monthly['Cluster']==cid]
         .set_index('date')
         .asfreq('M'))   

y = sub['total_amount'].astype(float)

# stationarity suggestions
m = 12  # monthly seasonality
p_raw, p_d1, p_seas = adf_p(y), adf_p(y.diff()), adf_p(y.diff(m))
d = 0 if p_raw < 0.05 else (1 if p_d1 < 0.05 else 2)
D = 1 if p_seas >= 0.05 else 0

# split
train_ratio = 0.8
n = len(y); n_tr = int(n*train_ratio)
y_tr, y_te = y.iloc[:n_tr], y.iloc[n_tr:]

# choose orders 
order = (1, d, 1)
seasonal_order = (1, D, 1, m)


# fit
res = SARIMAX(y_tr, order=order, seasonal_order=seasonal_order,
              enforce_stationarity=False, enforce_invertibility=False).fit(disp=False)

# test forecast + metrics
pred = res.get_prediction(start=y_te.index[0], end=y_te.index[-1], dynamic=False)
y_hat = pred.predicted_mean
print("Cluster", cid, "orders:", order, seasonal_order)
print(metrics(y_te, y_hat))

```

### SARIMA for ALL CLUSTERS

```python
from statsmodels.stats.diagnostic import acorr_ljungbox

def fit_cluster_df(grp, m=12, train_ratio=0.8):
    s = grp.set_index('date').asfreq('M')['total_amount'].astype(float)

    # differencing hints
    p_raw, p_d1, p_seas = adf_p(s), adf_p(s.diff()), adf_p(s.diff(m))
    d = 0 if p_raw < 0.05 else (1 if p_d1 < 0.05 else 2)
    D = 1 if p_seas >= 0.05 else 0

    n=len(s); n_tr=int(n*train_ratio)
    y_tr, y_te = s.iloc[:n_tr], s.iloc[n_tr:]

    order=(1,d,1); seasonal=(1,D,1,m)
    

    res = SARIMAX(y_tr, order=order, seasonal_order=seasonal,
                  enforce_stationarity=False, enforce_invertibility=False).fit(disp=False)

    if len(y_te)==0:
        return order, seasonal, {}, (np.nan, np.nan)

    pred = res.get_prediction(start=y_te.index[0], end=y_te.index[-1], dynamic=False)
    y_hat = pred.predicted_mean
    mets = metrics(y_te, y_hat)

    lb = acorr_ljungbox(res.resid.dropna(), lags=[12,24], return_df=True)
    lb12, lb24 = float(lb['lb_pvalue'].iloc[0]), float(lb['lb_pvalue'].iloc[-1])
    return order, seasonal, mets, (lb12, lb24)

rows=[]
for cid, grp in cluster_monthly.groupby('Cluster'):
    ord_, seas_, mets_, (lb12, lb24) = fit_cluster_df(grp, m=12, train_ratio=0.8)
    rows.append({"Cluster": cid, "order": ord_, "seasonal_order": seas_, **mets_,
                 "LB_p(12)": lb12, "LB_p(24)": lb24})

metrics_df = (
    pd.DataFrame(rows).sort_values("Cluster")
    .assign(**{
        "MAE":      lambda d: d["MAE"].round(2),
        "RMSE":     lambda d: d["RMSE"].round(2),
        "MAPE%":    lambda d: d["MAPE%"].round(3),
        "sMAPE%":   lambda d: d["sMAPE%"].round(3),
        "LB_p(12)": lambda d: d["LB_p(12)"].round(3),
        "LB_p(24)": lambda d: d["LB_p(24)"].round(3),
    })
)
metrics_df = pd.DataFrame(rows).sort_values("Cluster")
for c, r in {"MAE":2, "RMSE":2, "MAPE%":3, "sMAPE%":3, "LB_p(12)":3, "LB_p(24)":3}.items():
    if c in metrics_df.columns:
        metrics_df[c] = metrics_df[c].round(r)
metrics_df.to_csv("sarimax_cluster_metrics.csv", index=False)
```

### PREDICTING CLUSTER 3 next 6 months

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import numpy as np
import matplotlib.pyplot as plt

def forecast_cluster(cluster_id, H=6, m=12):
    sub = (cluster_monthly[cluster_monthly['Cluster']==cluster_id]
             .set_index('date').asfreq('M')['total_amount'].astype(float))

    # differencing hints
    from statsmodels.tsa.stattools import adfuller
    def adf_p(x):
        try: return adfuller(x.dropna(), autolag='AIC')[1]
        except: return np.nan
    p_raw, p_d1, p_seas = adf_p(sub), adf_p(sub.diff()), adf_p(sub.diff(m))
    d = 0 if p_raw < 0.05 else (1 if p_d1 < 0.05 else 2)
    D = 1 if p_seas >= 0.05 else 0

    order=(1,d,1); seasonal=(1,D,1,m)
    model = SARIMAX(sub, order=order, seasonal_order=seasonal,
                    enforce_stationarity=False, enforce_invertibility=False).fit(disp=False)

    fc = model.get_forecast(steps=H)
    out = fc.predicted_mean.rename('forecast').to_frame()
    out['lo'] = fc.conf_int().iloc[:,0]
    out['hi'] = fc.conf_int().iloc[:,1]

    # sanity: level consistency
    last_hist = float(sub.iloc[-1])
    first_fc  = float(out['forecast'].iloc[0])
    print(f"Cluster {cluster_id}: last history={last_hist:,.0f}, first forecast={first_fc:,.0f}, "
          f"ratio={first_fc/last_hist:.3f}")

    # plot (zoomed to last 24 months)
    ax = sub.iloc[-24:].plot(figsize=(10,4), label="History", lw=1.2, color="tab:pink")
    out['forecast'].plot(ax=ax, label="Forecast", color="tab:orange", lw=1.8)
    plt.fill_between(out.index, out['lo'], out['hi'], alpha=0.2, label="95% PI")
    plt.axvline(sub.index[-1], color='gray', ls='--', lw=1)
    ax.set_title(f"Cluster {cluster_id} — {H}-Month Forecast (order={order}, seasonal={seasonal})")
    ax.set_xlabel("date"); ax.set_ylabel("total_amount"); ax.legend()
    plt.tight_layout(); plt.show()
    return sub, out

# Recompute for the cluster you’re plotting (e.g., 3)
_ = forecast_cluster(3, H=6, m=12)
```
